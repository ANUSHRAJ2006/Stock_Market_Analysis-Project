#HEAT MAP OF MONTHLY RETURN

```python
tech_tickers=['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META']
tech_df=df[df['Ticker'].isin(tech_tickers) & (df['Year']== 2023)]
tech_df.head()
```

```python
# Calculate monthly returns for each stock
result_data = []

for ticker in tech_tickers:
    ticker_data = tech_df[tech_df['Ticker'] == ticker].copy()

    # Group by month
    monthly_groups = ticker_data.groupby(tech_df['Date'].dt.month)

    for month_num, group in monthly_groups:
        month_name = group['Date'].iloc[0].strftime('%b')

        # Get first and last trading day of month
        start_price = group.iloc[0]['Open']
        end_price = group.iloc[-1]['Close']

        # Calculate return
        monthly_return = ((end_price - start_price) / start_price) * 100

        result_data.append({'Ticker': ticker,
                            'Month': month_name,
                            'Month_num': month_num,  # Store month number for sorting
                            'Return': monthly_return})

# Convert to DataFrame
monthly_df = pd.DataFrame(result_data)
monthly_df.head()
```

```python
pivot_df=monthly_df.pivot(index='Month_num',columns='Ticker',values='Return')
pivot_df.sort_index(inplace=True)
pivot_df
```

```python
month_labels = [pd.Timestamp(2023,month,1).strftime('%b') for month in pivot_df.index]
fig=go.Figure(data=go.Heatmap(z=pivot_df.values,
                              x=pivot_df.columns,
                              y=month_labels,
                              colorscale=[[0,'maroon'],[0.5,'white'],[1,'green']],
                              zmid=0,
                              colorbar=dict(
                                  title='Monthly Return (%)',
                                  titleside='right'),
                              text=np.round(pivot_df.values,2),
                              texttemplate='%{text:.2f}%',
                              textfont={'size':12}))
fig.update_layout(title='Monthly Returns of 5  Tech stocks (2023)',
                  xaxis_title='Stock Ticker',
                  yaxis_title='Month',
                  height=600,
                  width=1200,
                  xaxis={'side':'top'})
fig.show()
```

#VOLUME GROUPED BAR CHART

```python
tech_tickers=['AAPL','MSFT','GOOGL']
tech_names={'AAPL':'Apple','MSFT':'Microsoft','GOOGL':'Google'}
tech_df=df[(df['Ticker'].isin(tech_tickers)) &
         (df['Year']>=2020) & (df['Year']<=2025)].sort_values('Date')
tech_df[['Date','Ticker','Volume']].head()
```

```python
yearly_volume=tech_df.groupby(['Year','Ticker'])['Volume'].sum().reset_index()
yearly_volume['Volume_Billions']=yearly_volume['Volume']/1_000_000_000
```

```python
fig=go.Figure()

for i,ticker in enumerate(tech_tickers):
    company_data=yearly_volume[yearly_volume['Ticker']==ticker]
    company_data

    fig.add_trace(go.Bar(x=company_data['Year'],
                         y=company_data['Volume_Billions'],
                         name=tech_names[ticker]))


fig.update_layout(title='Annual trading : Tech Coampanies (2020-2025)',
                   xaxis_title='Year',
                   yaxis_title='Volume(Billions)',
                   template='plotly_white',
                   barmode='group',
                   height=700,
                   width=800,
                   legend=dict(orientation='h',y=1.1,x=0.7)
                   )
fig.show()
```
